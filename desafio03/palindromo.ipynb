```python
def palindromic_numbers(start: int, end: int) -> list[int]:
    if start <= 0 or end <= 0:
        raise ValueError("Apenas inteiros positivos são permitidos")

    if start > end:
        start, end = end, start

    result = []
    for num in range(start, end + 1):
        s = str(num)
        if s == s[::-1]:
            result.append(num)

    return result


# Testes pedidos no desafio
print(palindromic_numbers(1, 20))
print(palindromic_numbers(3000, 3010))
print(palindromic_numbers(101, 121))
```
